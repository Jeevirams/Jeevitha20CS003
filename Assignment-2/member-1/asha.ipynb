```python
# import all the necessery libraries

import pandas as pd
import numpy as np
```

```python
# read the dataset

df = pd.read_csv('/content/drive/MyDrive/AI/House Price India.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# split the data(independent & dependent)

x = df.iloc[:,0:-1].values
y= df.iloc[:,4:5].values
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.3,random_state = 0)
```

```python

xtrain.shape
```

```python

xtrain.shape
```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python

# create ANN structure

regressor = Sequential()
regressor.add(Dense(4,activation = 'relu'))
regressor.add(Dense(12,activation = 'relu'))
regressor.add(Dense(9,activation = 'relu'))
regressor.add(Dense(1,activation = 'linear'))
```

```python


# compiling the ANN model
 
regressor.compile(optimizer = 'adam',loss= 'mse',metrics = ['mse'])
```

```python
# Training the model
regressor.fit(xtrain,ytrain,batch_size = 10,epochs = 300)
```

```python
ytest
```

```python

y_pred = regressor.predict(xtest)
y_pred
```

```python
# evaluation

from sklearn.metrics import r2_score

r2_score(ytest,y_pred)*100
```
